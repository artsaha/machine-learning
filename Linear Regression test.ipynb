```python
from sklearn.model_selection import train_test_split 
from sklearn import datasets 
import matplotlib.pyplot as plt 

X,y = datasets.make_regression(n_samples = 132, n_features = 1, noise = 18, random_state=23)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

```python
fig = plt.figure(figsize = (8,6))
plt.scatter(X[:,0], y, color="orange", marker = "o", s = 30)
plt.show()
```

```python
from LinReg_impl import LinReg 
from LinReg import LinearReg
reg = LinReg() 
reg.fit(X_train, y_train)
```

```python
y_hat = reg.predict(X_test)
y_hat
```

```python
import numpy as np 
def mean_square_error(y_pred, y_true):
    N = y_pred.shape[0]
    total = 0 
    for i in range(N):
        total += (y_pred[i] - y_true[i])**2

    return total / N 


def mse(y_pred, y_true):
    return np.mean((y_pred - y_true)**2)
```

```python
error1 = mse(y_hat, y_test)
print(error1)
error2 = mean_square_error(y_hat, y_test)
print(error2)
```
